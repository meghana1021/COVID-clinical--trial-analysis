## 2. Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../dataset/covid_clinical_trials.csv')

df.head()
```

## 3. Dataset Information

```python

print("Dataset Shape:", df.shape)

df.info()
```

## 4. Missing Values Analysis

```python

missing_values = df.isnull().sum()

missing_values[missing_values > 0]
```


## 5. Data Cleaning

### Cleaning Steps
- Handled missing values
- Standardized column names
- Converted date columns
- Prepared data for analysis


```python

df.columns = [col.strip().replace(" ", "_") for col in df.columns]

date_columns = ['Start_Date', 'Completion_Date', 'First_Posted']

for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

df.head()
```

## 6. Statistical Summary

```python

df.describe(include='all')
```

## 7. KPI Analysis

```python

total_trials = len(df)

study_types = df['Study_Type'].nunique() if 'Study_Type' in df.columns else 0

conditions = df['Conditions'].nunique() if 'Conditions' in df.columns else 0

participants = df['Enrollment'].sum() if 'Enrollment' in df.columns else 0

print("Total Clinical Trials:", total_trials)
print("Total Study Types:", study_types)
print("Total Conditions Studied:", conditions)
print("Total Participants:", participants)
```

# 8. Exploratory Data Analysis & Charts

## Chart 1: Clinical Trial Status Distribution

```python

if 'Status' in df.columns:
    plt.figure(figsize=(10,5))

    df['Status'].value_counts().head(10).plot(kind='bar')

    plt.title('Distribution of Clinical Trial Status')
    plt.xlabel('Status')
    plt.ylabel('Count')
    plt.xticks(rotation=45)

    plt.show()
```

## Chart 2: Clinical Trials by Study Type

```python

if 'Study_Type' in df.columns:
    plt.figure(figsize=(8,5))

    df['Study_Type'].value_counts().plot(kind='bar')

    plt.title('Clinical Trials by Study Type')
    plt.xlabel('Study Type')
    plt.ylabel('Count')

    plt.show()
```

## Chart 3: Top 10 Conditions Studied

```python

if 'Conditions' in df.columns:
    plt.figure(figsize=(10,6))

    df['Conditions'].value_counts().head(10).plot(kind='barh')

    plt.title('Top 10 Conditions Studied')
    plt.xlabel('Count')

    plt.show()
```

## Chart 4: Clinical Trials Trend Over Time

```python

if 'Start_Date' in df.columns:
    trend = df['Start_Date'].dt.year.value_counts().sort_index()

    plt.figure(figsize=(10,5))

    trend.plot(kind='line', marker='o')

    plt.title('Clinical Trials Over Time')
    plt.xlabel('Year')
    plt.ylabel('Number of Trials')
    plt.grid(True)

    plt.show()
```


# 9. Key Insights

### Insights
- Recruiting and Completed studies dominate the dataset.
- Interventional trials are the most common study type.
- COVID-19 and SARS-CoV-2 are the top researched conditions.
